```python
from langchain.document_loaders import UnstructuredURLLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Pinecone
import pinecone
from langchain.embeddings import HuggingFaceEmbeddings
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
from langchain import HuggingFacePipeline
from huggingface_hub import notebook_login
import torch
```

## Extract data from url

```python
!pip install numpy==1.24.4
!pip install scipy
```

```python
URLs = [
    "https://ineuron.ai/"
]
```

```python
loader = UnstructuredURLLoader(urls = URLs)
data = loader.load()
```

```python
data
```

## Chunkins

```python
text_splitter=CharacterTextSplitter(separator='\n',
                                    chunk_size=1000,
                                    chunk_overlap=200)
```

```python
text_chunks=text_splitter.split_documents(data)
```

```python
len(text_chunks)
```

```python
text_chunks[0]
```

## Embedding model

```python
embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
query_result = embeddings.embed_query("Hello world")
len(query_result)
```

```python
query_result
```

```python
import os

PINECONE_API_KEY = os.environ.get('PINECONE_API_KEY', '5e1c6876-5b76-499d-a3ec-002764f6887e')
PINECONE_API_ENV = os.environ.get('PINECONE_API_ENV', 'gcp-starter')
```

```python
import pinecone
# initialize pinecone
pinecone.init(
    api_key=PINECONE_API_KEY,  # find at app.pinecone.io
    environment=PINECONE_API_ENV  # next to api key in console
)
index_name = "website-bot" # put in the name of your pinecone index here
```

```python
docsearch = Pinecone.from_texts([t.page_content for t in text_chunks], embeddings, index_name=index_name)
```

## Create LLM wrapper

```python
notebook_login()
```

```python
model="meta-llama/Llama-2-7b-chat-hf"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model)
```

```python
pipe = transformers.pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm.invoke("Please provide a concise summary of the Book Harry Potter")
```

## Initialize the Retrieval QA

```python
from langchain.chains import RetrievalQA
```

```python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=docsearch.as_retriever())
```

```python
query = "Tell me the price of Full Stack Data Science course with Generative AI provided by ineuron"
```

```python
print(qa.run(query))
```
